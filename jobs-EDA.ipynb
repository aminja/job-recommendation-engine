### Exploratory data analysis of job positions

This phase investigates the structure and features of job's data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

jobs = pd.read_csv('data/Combined_Jobs_Final.csv')

jobs.head(5)
```

```python
# check overall view of columns

jobs.info()
```

```python
plt.figure(figsize=(16,4))
jobs['Education.Required'].value_counts().plot.bar()
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.title("Required Education for The Jobs")
plt.show()
```
